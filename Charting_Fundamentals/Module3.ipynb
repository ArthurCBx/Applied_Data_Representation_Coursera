# Histograms

```python
fig,((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2, figsize=(12,4),sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0, len(axs)):
  sample_size = 10**(n+1)
  sample = np.random.normal(loc=0.0, scale=1, size=sample_size)
  axs[n].hist(sample,bins=100) # bins is the number of bars to be plot
  axs[n].set_title(f'n={sample_size}')
```

```python
import matplotlib.gridspec as gridspec
plt.figure(figsize=(10,10))
gspec = gridspec.GridSpec(3,3)

# The GridSpec is indexed as rows and columns using the indexing operator, or square brackets.
# We'll create the first subplot in row0, covering off the first element, the middle position,
# and going to the end of the row.
top_histogram = plt.subplot(gspec[0,1:])

# We'll create the side histogram starting in row1 and continuing to the end of row elements,
# but limited to the column0
side_histogram = plt.subplot(gspec[1:,0])

# Finally, we'll create the scatter plot in the lower right. When we execute this, we see our basic set
# of plots
lower_right = plt.subplot(gspec[1:,1:])
```

```python
X = np.random.random(size=10000)
Y = np.random.normal(loc=0.0, scale=1, size=10000)

plt.figure(figsize=(10,10))
gspec = gridspec.GridSpec(3,3)

top_histogram = plt.subplot(gspec[0,1:])
side_histogram = plt.subplot(gspec[1:,0])
lower_right = plt.subplot(gspec[1:,1:])

lower_right.scatter(X,Y)
top_histogram.hist(X, bins=100)
side_histogram.hist(Y,bins=100,orientation='horizontal');
```

```python
X = np.random.random(size=10000)
Y = np.random.normal(loc=0.0, scale=1, size=10000)

plt.figure(figsize=(10,10))
gspec = gridspec.GridSpec(3,3)

top_histogram = plt.subplot(gspec[0,1:])
side_histogram = plt.subplot(gspec[1:,0])
lower_right = plt.subplot(gspec[1:,1:])

lower_right.scatter(X,Y)
top_histogram.hist(X, bins=100,density=True)
side_histogram.hist(Y,bins=100,orientation='horizontal',density=True);

# Rotate 180
side_histogram.invert_xaxis()

for ax in [top_histogram, lower_right]:
  ax.set_xlim(0,1)

for ax in [side_histogram, lower_right]:
  ax.set_ylim(-5,5)
```

# Animation

```python
import matplotlib.animation as animation

n = 100
x = np.random.randn(n)

def update(curr):
  if curr == n:
    # If we're in the last frame, stop the animation
    anim.event_source.stop()

  # Clear the current axes
  plt.cla()

  bins=np.arange(-4,4,0.5)

  plt.hist(x[:curr], bins=bins)

  plt.axis([-4,4,0,30])

  plt.gca().set_title('Sampling the Normal Distribution')
  plt.gca().set_ylabel('Frequency')
  plt.gca().set_xlabel('Value')
  plt.annotate(f'n={curr, [3,27]}')
```

```python
anim = animation.FuncAnimation(plt.figure(), update, interval=100, cache_frame_data=False)

plt.show()
```

# Widget Demonstration

```python
from ipywidgets import interact
import ipywidgets as widgets

df = pd.read_csv('Applied_Data_Representation_Coursera/Module3/wipeout.csv')
df = df.set_index(pd.to_datetime(df['timestamp']))
df.head()
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
potential_columns=[]
for col in df.columns:
  if df[col].dtype != "object":
    potential_columns.append(col)
```

```python
def plot_data(col_names):
  plt.figure(figsize=[12,8])
  plt.gcf().suptitle(str(col_names))
  if len(col_names)==2:
    plt.plot(df[list(col_names)[0]],'b')
    ax=plt.twinx() # Guarantee the x-axis is the same, but y-axis can be unique
    ax.plot(df[list(col_names)[1]],'r')
  else:
    plt.plot(df[list(col_names)])
```

```python
plot_data(['cadence','enhanced_speed'])
```

```python
@interact(col_names=widgets.SelectMultiple(
    options=potential_columns, # The column names
    value=(potential_columns[0],), # Its time followed by any number of other items selected
    description='Columns'))
def plot_data(col_names): # The same function
    plt.figure(figsize=[12,8])
    plt.gcf().suptitle(str(col_names))
    if len(col_names) ==2:
        plt.plot(df[list(col_names)[0]],'b')
        ax=plt.twinx()
        ax.plot(df[list(col_names)[1]],'r')
    else:
        plt.plot(df[list(col_names)])
```

```python
@interact(heart_rate_bounds=widgets.IntRangeSlider(
    value=[80, 100],
    min=0,
    max=180,
    step=1,
    description='HR Range:',
    continuous_update=False,
    readout_format='d',),
    start=widgets.Text(
    value=str(df.index[0]),
    description='Start Time:'),
    end=widgets.Text(
    value=str(df.index[-1]),
    description='End Time:'))

def plot_hr(heart_rate_bounds, start, end):
    plt.figure(figsize=[12,8])

    plt.xlim(left=df.index[0],right=df.index[-1])
    plt.ylim(0,max(df["heart_rate"]))

    ndf=df[(df["heart_rate"]>heart_rate_bounds[0]) &
           (df["heart_rate"]<heart_rate_bounds[1]) &
           (df.index>start) & (df.index<end)]
    plt.scatter(ndf.index,ndf["heart_rate"])
```
