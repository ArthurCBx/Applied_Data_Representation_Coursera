In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650),
                   np.random.normal(43000,100000,3650),
                   np.random.normal(43500,140000,3650),
                   np.random.normal(48000,70000,3650)],
                  index=[1992,1993,1994,1995])
df
```

```python
%matplotlib widget
#use cell magic to make widget work

from matplotlib.widgets import Cursor
import matplotlib.pyplot as plt

#convert index to string for better chart labelling
df.index=df.index.astype(str)

#compute the mean for each year
s_mean=df.mean(axis=1)
#s_mean #used to check data

#compute the standard error of means
s_sem=df.sem(axis=1)
#s_sem #used to check data

#get blue-white-red color map with x intervals (discrete instead of conti
my_cmap=plt.get_cmap("bwr",11)
#my_cmap #used to check color map

#set starting values for variables
enabled=False
my_colors=["r"]
my_scale=[]

#create the function which allows the usage of a cursor
#the cursor can be pinned and unpinned using the left mouse button
#additionnaly depending on the cursor position the bar colors are recalculated
def onevent(event):
    global enabled
    global my_colors
    global my_scale
    global my_text
    global my_line

    #checks for left mous click
    if event.button==1:
        if enabled==True:
            #cursor is activated aka visibly following the mouse pointer
            cursor.set_active(True)

            #remove previously printed text and line
            my_text.remove()
            my_line.remove()

            #reset toggle variable
            enabled=False
        else:
            #cursor is deactivated aka not visibly following the mouse pointer
            cursor.set_active(False)
            y=event.ydata

            #print line at the current cursor position including a textbox with the y value
            my_text=plt.text(-0.55, y+1100, f"y = {y:.0f}",fontsize=11,bbox=dict(facecolor="silver"))
            my_line=plt.axhline(y, color="k", linewidth=2,linestyle="--")

            #recalculate values for the colors depending on the cursor position
            #clear the content of scale befor calculating new values
            my_scale=[]
            for i in range(len(s_mean)):
                #when cursor is below the range of the data bar the color should be red => value 1
                if y < (s_mean[i]-s_sem[i]):
                    my_scale.append(1)
                #when cursor is above the range of the data bar the color should be blue => value 0
                elif y > (s_mean[i]+s_sem[i]):
                    my_scale.append(0)
                #when the cursor is somewhere in the range of the data the color values should be between 0 and 1
                else:
                    my_scale.append((abs(y-(s_mean[i]+s_sem[i])))/(s_sem[i]*2))
            #convert the values in the scale to float, else the color mapping isn´t working properly (issue with 1.0)
            my_scale=[float(x) for x in my_scale]

            #change the colors based on the new my_scale
            my_colors=[]
            for i in my_scale:
                my_colors.append(my_cmap(i))

            #plot data again with new colors
            plt.bar(s_mean.index,s_mean.values,color=my_colors,edgecolor="k")

            #reset toggle variable
            enabled=True


#create figure with custom size
fig, ax=plt.subplots(figsize=(10,10))

#create a plot for the colorbar legend which will never be shown
plot=plt.scatter(my_scale,my_scale,c=my_scale,cmap=my_cmap)
plt.cla()

#plot data bars based on mean values
plt.bar(s_mean.index,s_mean.values,color=my_colors,edgecolor="k")

#fix x-axis to avoid resacling after mouse clicks
ax.set_xlim([-0.6,3.6])

#plot error bars based on standard error of means
plt.errorbar(s_mean.index,s_mean.values,yerr=s_sem.values,fmt="none",color="k",capsize=20)

#add colorbar legended
plt.colorbar(plot,ticks=[],location="bottom",label="Color scale\nBlue: data is certainly below cursor value\nWhite: data certainly contains cursor value\nRed: data is certainly above cursor value")

#include cursor#
cursor=Cursor(ax,vertOn=False,useblit=True,color="k",linewidth=2,linestyle="--")
fig.canvas.mpl_connect("button_press_event", onevent)

#Enhance visual
plt.xlabel("Year",fontsize=11)
plt.ylabel("Data",fontsize=11)
plt.title("Bar chart of data with standard error of means from 1992 through 1995\n\nHint: using the left mouse button the cursor can be pinned or unpinned",fontsize=14);
```
