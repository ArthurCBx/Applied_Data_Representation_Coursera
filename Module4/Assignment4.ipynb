```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'year': [year for year in range(2011,2026,1)],
    'worlds prize (millions)': [0.100, 2, 2.05, 2.13, 2.13, 5.07, 4.947, 6.45, 2.225, 2.34, 2.18, 2.225, 2.225, 2.225, np.nan],
    'active players (millions)': [15, 32, 67, 70, 90, 100, 81, 75, 116, 137, 149, 152, 152, 132, 131],
    'peak viewers excluding China (millions)': [np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,2.102,2.040,3.981,3.884,4.018,5.147,6.403,6.857,np.nan]
})
df
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(8,8))

ax1.set_xlabel('Year')
ax1.set_ylabel('Active players (millions)',alpha = 0.8)
ax1.plot(df['year'], df['active players (millions)'],'-', color='black',linewidth=1.5,alpha=0.7)
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
ax2.set_ylabel('Worlds prize & Peak viewers (millions)',alpha=0.8)
ax2.plot(df['year'], df['worlds prize (millions)'],'--',color='gray',linewidth=1.5)
ax2.tick_params(axis='y')
ax2.plot(df['year'], df['peak viewers excluding China (millions)'],':' ,color='gray',linewidth=1.5)

for spine in ax1.spines.values():
    spine.set_visible(False)

for spine in ax2.spines.values():
    spine.set_visible(False)

fig.legend(['Active players', 'Worlds prize', 'Peak viewers (excluding China)'],
           bbox_to_anchor=(0.5,0.2), frameon=False, fontsize=9)
ax1.set_title('League of Legends Worlds prize vs Active player\n and Peak Worlds viewers over the years',fontsize=12,alpha=0.9)


fig.tight_layout()
plt.savefig('Lol_stats.png')
```
